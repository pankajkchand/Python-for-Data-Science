##### Matplotlib
* It is a popular Python library used for creating static, interactive, and animated visualization.
* It is often used for generating plots, charts, histograms, scatterplots and other graphical represantations of data.
* Matplotlib provides a wide variety of customization options allowing users to create publication-quality visualization in a simple and   straightforward manner

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Lineplot
* Lineplot is commonly used to visualize data trends over a continuous variable such as time.

```python
x=np.linspace(4,15,100)
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label="Sin Wave",linestyle="-",linewidth=2,color="blue")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")

plt.grid(True)
plt.show()
```

### Barchart
* Effective for comparing categorical data
* alpha--> for transparency ,ranges from 0 to 1

```python
category=["A","B","C","D","E"]
value=[33,54,42,51,29]
```

```python
plt.figure(figsize=(5,3))
plt.bar(category,value,alpha=0.7,edgecolor="black",color=["#234269","#234321","#234290","#234476","#234250"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Categories")
plt.grid(True)
plt.show()
```

### Scatter Plot
* Helps visualize the relationship between two variables(Numeric Values).

```python
x=np.random.rand(100)
y=np.random.rand(100)
color=np.random.rand(100)
size=1000*np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x, y, c=color, s=size, alpha=0.5, cmap="plasma")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.colorbar(label="Color Intensity")
plt.grid(True)
plt.show()
```

### Histogram
* Displays the distribution of a single numeric variable.

```python
dice = np.array([1, 2, 3, 4, 5, 6]) # Dice
sample = np.random.choice(dice, 4) # Rolling dice 4 time in a row and getting values
print("Output of 4 samples", sample)
print("Mean of above 4 samples", sample.mean())
```

```python
samples_1000 = [np.random.choice(dice, 4).mean() for i in range(1000)]
samples_1000
```

```python
plt.figure(figsize=(5,3))
plt.hist(samples_1000, alpha=0.7, edgecolor="black", bins=100, histtype="barstacked")
plt.show()
```

```python
data=np.random.randn(10000)
```

```python
plt.figure(figsize=(5,3))
plt.hist(data,alpha=0.7,edgecolor="black",bins=100,histtype="stepfilled",color="r")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

### Pie Chart
* Pie Chart represents parts of a whole.
* explode--> Emphasizes a particular segment by pulling it out.

```python
sizes=np.random.randint(35,89,4)
labels=["East","West","North","South"]
```

```python
plt.figure(figsize=(5,4))
plt.pie(sizes,labels=labels,colors=["gold", "lightgreen", "lightblue", "lightcoral"],autopct="%1.1f%%",explode=[0.01,0.1,0.01,0.01])
plt.title("Pie Chart")
plt.show()
```

### Boxplot
* Useful for visualizing the distribution of a dataset through 5 summary statistics: Minimum, First Quartile, Median, Third Quartile and Maximum

```python
data=np.random.normal(loc=0,scale=1,size=(100,))

plt.figure(figsize=(5,4))
plt.boxplot(data,patch_artist=True,vert=False)
plt.title("Boxplot")
plt.show()
```

### Polar Plot
* A Polar Plot represents data in a circular format, commonly used for visualising cyclic patterns or directional data.

```python
angles=np.linspace(0,2*np.pi,100)
radii=np.random.rand(100)

plt.figure(figsize=(5,4))
ax=plt.subplot(111,polar=True)
ax.plot(angles,radii,color="blue",linewidth=0.2)
ax.fill(angles,radii,color="blue",alpha=0.3)
plt.title("Polar Plot")
plt.show()
```

### Heatmap
* Graphical representations of data where value in a matrix are represented as colors.

```python
matrix = np.random.rand(5,5)
plt.figure(figsize=(5, 4))
plt.imshow(matrix, cmap="plasma")
plt.title("Heatmap")
plt.colorbar(label="Color Intensity")
plt.show()
```

```python
overs = np.arange(1, 21)
ind = np.random.randint(1, 24, 20)
nz = np.random.randint(1, 24, 20)
```

```python
ind_total = [0]
for i in ind:
    ind_total.append(i+ind_total[-1])

nz_total = [0]
for i in nz:
    nz_total.append(i+nz_total[-1])
```

```python
plt.figure(figsize=(10, 3))

plt.subplot(121)
plt.plot(overs, nz_total[1:], label="New Zealand")
plt.plot(overs, ind_total[1:], label="India")
plt.legend()

plt.subplot(122)
plt.bar(overs, nz, label="New Zealand")
plt.bar(overs, ind, label="India")
plt.legend()
plt.show()
```
