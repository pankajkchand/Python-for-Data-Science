# Seaborn
* Python data visualization library
* Easily creates the most common types of plots.
* Advance version of Matplotlib.


#### Why It is useful?
* Gather data-->Transform and Clean-->Explore-->Analyse & Build Models-->Communicate Results

#### Advantages
* Easy to use.
* Works well with Pandas data structures.
* Built on top of Matplotlib. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
# from matplotlin import pyplot as plt
```

### Scatterplot

```python
weight = [62,64,69,75,66,68,65,71,76,73]
height = [120,136,148,175,137,165,154,172,200,187]
```

```python
%matplotlib inline
plt.figure(figsize=(4.5,2.5))   # size can be given in tuple
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
sns.scatterplot(x=height, y=weight)
plt.show()
```

### Countplot

```python
gender=["Female","Female","Female","Female","Male","Male","Male","Male","Male","Male"]
```

```python
plt.figure(figsize=(4,2.5))
plt.title("Count of Female and Male")
sns.countplot(x=gender,hue=gender,legend=["female","Male"],fill=False)
plt.show()
```

### Using Pandas with Seaborn

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("SampleSuperstore.xlsx")
df.head()
```

```python
sns.countplot(data=df, x="Ship Mode", hue="Segment")
plt.show()
```

```python
df.columns
```

```python
sns.scatterplot(data=df, x="Quantity", y="Sales", hue="Category") 
plt.show()
```

```python
plt.figure(figsize=(20,5))   # 15= width, 5= height
sns.histplot(data=df, x="Sub-Category")
plt.xticks(rotation=60)  # for items name in X-Axis 
plt.show()
```

```python
sns.scatterplot(data=df, x="Quantity", y="Sales", hue="Category", 
                hue_order=["Technology","Office Supplies","Furniture"],
               palette={"Technology":"red","Office Supplies":"blue","Furniture":"black"})
plt.show()
```
